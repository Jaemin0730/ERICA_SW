```python
## 문제
## 시행횟수에 따른 동전의 앞면이 나올 확률의 변화값을 선그래프로 그리고, 
## 1000번째의 확률값을 제목에 소숫점 4째자리에서 반올림해서 나타내시오.
## (단, 총 시행횟수는 1000 번 시행하고, while 문을 이용하시오.)
import numpy as np
import pandas as pd
import random
import matplotlib.pyplot as plt
```

```python
h = 0
n = 0
arrx = []
arry = []
coin = ['H', 'T']
while (n < 1000):
    choice = random.choice(coin)
    if choice == 'H':
        h += 1
    n += 1
    c = h/n
    arrx.append(n)
    arry.append(c)

print('Coin : P(H) = ', round(h/n,3))
#print(arrx)
#print(arry)
```

```python
x = arrx
y = arry
k = str(round(h/n,3))
plt.plot(x, y, 'bo--')
plt.title('Coin : P(H) = ' + k)
plt.xlabel("Times")
plt.ylabel("Prob")
```
