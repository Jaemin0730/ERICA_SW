```python
# 열 새로 추가 -> 'density'
# -> 데이터파일명['추가하고 싶은 이름'] = 넣을 값 으로 만든다~~!~!
country2['density'] = country2['population']/country2['area']
country2
```

```python
# 새로운 행 '프랑스' 데이터를 추가하기
# pd.DataFrame()을 사용하여, (index=['행이름'], data=[('값','값', ....)], columns=[])을 넣는다
# data: 열에 들어갈 값들, columns: 열의 이름들

# 추가하는법: 데이터파일명.append()를 사용하자!
new_value = pd.DataFrame(index = ['FR'], data = [('France', 265449, 'Paris', 126793004, 344567)], 
                            columns = ['country', 'area', 'capital', 'population', 'density'])

country3 = country2.append(new_value)
country3
```

```python
# 관찰에 필터 설정하기(조건 설정) -> [여기에 조건 입력]
country3[country3['population'] > 20000000]
```

```python
# NaN 데이터를 찾기 (in country3)
# -> .isna()를 사용하면 nan인 데이터는 True, nan이 아닌 데이터는 False로 보인다.
country3.isna()
```

```python
# NaN 데이터를 가진 행 삭제하기
# dropna(axis=0, how='any', inplace=False)
# axis = 0:row(행), 1:column(열)
#       -> axis가 0,1인지에 따라 nan이 있는 행,열이 모두 지워진다
# how = 'any': NaN이 row나 column에 하나라도 있으면 삭제
#       'all': NaN이 row나 column에 모두 있어야지 삭제
# inplace = 'True' : dropna가 적용된 dataframe을 기존의 dataframe에 적용
#           'False': dropna가 적용된 dataframe과 기존의 dataframe은 독립

country4 = country3.dropna(axis=0, how='any', inplace=False)
# 여기서는 inplace가 false이기에 country3에는 영향이 없고, country4에만 dropna가 적용!

country4
```

```python
# NaN 데이터를 0으로 바꾸기
# -> .fillna()
# inplace = 'True' : fillna가 적용된 dataframe을 기존의 dataframe에 적용
#           'False': fillna가 적용된 dataframe과 기존의 dataframe은 독립

country3.fillna(0, inplace=True)
# 여기서는 inplace가 true이기에 country3에는 바로 영향을 준다!
country3
```
